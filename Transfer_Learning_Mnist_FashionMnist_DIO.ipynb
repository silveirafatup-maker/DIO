Criar e treinar um modelo com o dataset MNIST

```python
import numpy as np

import tensorflow as tf

from tensorflow import keras

from tensorflow.keras import layers

import matplotlib.pyplot as plt
```

```python
# Carregar o dataset MNIST

mnist = keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# Preprocessar os dados

x_train = x_train.astype("float32") / 255.0

x_test = x_test.astype("float32") / 255.0
```

```python
# Definir o modelo

model_mnist = keras.Sequential([

 layers.Flatten(input_shape=(28, 28)),

 layers.Dense(128, activation='relu'),

 layers.Dropout(0.2),

 layers.Dense(10, activation='softmax')

])
```

```python
# Exibir dados... conferência...

plt.figure(figsize=(10,10))
for i in range(10):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i], cmap=plt.cm.binary)
plt.show()
```

```python
# Compilar o modelo

model_mnist.compile(optimizer='adam',

 loss='sparse_categorical_crossentropy',

 metrics=['accuracy'])
```

```python
# Treinar o modelo

model_mnist.fit(x_train, y_train, epochs=10)

```

```python
# Avaliar o modelo

test_loss, test_acc = model_mnist.evaluate(x_test, y_test, verbose=2)

print(f'Acurácia do modelo MNIST: {test_acc}')

```

## Preparar o dataset Fashion MNIST

### Subtask:
Carregar, pré-processar e dividir os dados do Fashion MNIST em conjuntos de treino e teste.

```python
# Carregar o dataset Fashion MNIST
fashion_mnist = keras.datasets.fashion_mnist

(x_train_fashion, y_train_fashion), (x_test_fashion, y_test_fashion) = fashion_mnist.load_data()

# Preprocessar os dados (normalizar)
x_train_fashion = x_train_fashion.astype("float32") / 255.0
x_test_fashion = x_test_fashion.astype("float32") / 255.0

# Opcional: Verificar o formato dos dados
print("Formato dos dados de treino do Fashion MNIST:", x_train_fashion.shape)
print("Formato dos dados de teste do Fashion MNIST:", x_test_fashion.shape)
```

## Combinar os modelos

### Subtask:
Combinar os modelos treinados em MNIST e Fashion MNIST usando a média das previsões.

```python
# Definir o modelo Fashion MNIST
model_fashion_mnist = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dropout(0.2), # Adicionando Dropout como no modelo MNIST
    keras.layers.Dense(10, activation='softmax') # 10 classes para Fashion MNIST
])

# Compilar o modelo
model_fashion_mnist.compile(optimizer='adam',
                            loss='sparse_categorical_crossentropy',
                            metrics=['accuracy'])

# Treinar o modelo
print("\nTreinando o modelo Fashion MNIST...")
model_fashion_mnist.fit(x_train_fashion, y_train_fashion, epochs=10)
```

## Carregar e Preparar o Modelo Pré-treinado para Transfer Learning

### Subtask:
Carregar um modelo pré-treinado e adaptá-lo para a classificação do CIFAR-10.

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Model

# Definir as dimensões da imagem de entrada (CIFAR-10 é 32x32x3)
input_shape = (32, 32, 3)
num_classes = 10

# Carregar o modelo VGG16 pré-treinado no ImageNet, excluindo a camada do topo
base_model = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)

# Adicionar novas camadas no topo do modelo base
x = base_model.output
x = Flatten()(x) # Achatando a saída do modelo base
x = Dense(256, activation='relu')(x) # Adicionando uma camada densa
x = Dropout(0.5)(x) # Adicionando Dropout para regularização
predictions = Dense(num_classes, activation='softmax')(x) # Camada de saída para 10 classes (CIFAR-10)

# Criar o modelo final
model_transfer_learning = Model(inputs=base_model.input, outputs=predictions)

# Congelar as camadas do modelo base (opcional, mas comum na primeira fase do Transfer Learning)
# Isso impede que os pesos das camadas pré-treinadas sejam atualizados durante o treinamento no novo dataset.
for layer in base_model.layers:
    layer.trainable = False

# Compilar o modelo
model_transfer_learning.compile(optimizer='adam',
                                loss='sparse_categorical_crossentropy', # Usamos sparse_categorical_crossentropy pois os rótulos são inteiros
                                metrics=['accuracy'])

# Mostrar um resumo do modelo
model_transfer_learning.summary()
```

```python
# Fazer previsões com o modelo MNIST nos dados de teste
predictions_mnist = model_mnist.predict(x_test)

# Fazer previsões com o modelo Fashion MNIST nos dados de teste
predictions_fashion_mnist = model_fashion_mnist.predict(x_test_fashion)

# Para combinar as previsões, precisamos garantir que os datasets de teste sejam compatíveis.
# Neste caso, como os datasets são diferentes, a combinação direta das previsões como um ensemble simples
# para um único problema de classificação não faz sentido.

# No entanto, se o objetivo fosse, por exemplo, um sistema que primeiro identifica se a imagem é um dígito ou uma peça de roupa
# e depois usa o modelo apropriado, a abordagem seria diferente.

# Dado o contexto atual de ter dois modelos treinados em datasets diferentes,
# uma forma de "combinar" seria usar a capacidade de cada modelo em seu respectivo domínio.
# Não é um ensemble no sentido tradicional para um único problema.

# Vamos avaliar cada modelo em seu próprio conjunto de teste e discutir como eles poderiam ser usados juntos
# em um cenário mais complexo.

print("\nAvaliação do modelo MNIST:")
test_loss_mnist, test_acc_mnist = model_mnist.evaluate(x_test, y_test, verbose=2)
print(f'Acurácia do modelo MNIST: {test_acc_mnist}')

print("\nAvaliação do modelo Fashion MNIST:")
test_loss_fashion, test_acc_fashion = model_fashion_mnist.evaluate(x_test_fashion, y_test_fashion, verbose=2)
print(f'Acurácia do modelo Fashion MNIST: {test_acc_fashion}')

# Documentação: Explicar por que a combinação direta por média não é aplicável aqui
# e discutir possíveis cenários de uso conjunto dos modelos.
```

3.  **Avaliação Individual**: Cada modelo foi avaliado em seu próprio conjunto de teste para verificar seu desempenho.
4.  **Discussão sobre a Combinação**: Foi observado que a combinação direta das previsões por média (um método comum de ensemble) não é aplicável neste caso, pois os modelos foram treinados em datasets com classes completamente diferentes (dígitos vs. roupas). Para um sistema que utilizasse ambos os modelos, seria necessário um mecanismo para determinar qual modelo usar para uma dada entrada (por exemplo, um classificador inicial para distinguir dígitos de roupas) ou uma arquitetura de modelo mais complexa projetada para lidar com ambos os tipos de dados simultaneamente.
