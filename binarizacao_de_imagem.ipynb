## Transformação de Imagem: Para Níveis de Cinza, depois para Binarizada (P&B)

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# URL da imagem fornecida de utilização gratuita
url_da_imagem = "https://img.freepik.com/fotos-gratis/jovem-linda-mulher-sorridente-em-roupas-da-moda-de-verao-mulher-despreocupada-sexy-posando-perto-da-parede-rosa-no-estudio-modelo-morena-positiva-se-divertindo-alegre-e-feliz-no-chapeu_158538-21565.jpg?semt=ais_incoming&w=740&q=80"
```

```python
# Baixar a imagem da internet
# verifica possíveis erros
try:
    resposta = requests.get(url_da_imagem)
    imagem_original = Image.open(BytesIO(resposta.content))
    print("Imagem baixada com sucesso!")
except requests.exceptions.RequestException as erro:
    print(f"Erro ao baixar a imagem. Verifique a conexão ou o link da imagem. Erro: {erro}")
    exit()    # Se der erro o programa termina
```

```python
# Imagem em níveis de Cinza
imagem_cinza = imagem_original.convert('L')
print("Imagem convertida para níveis de cinza.")
```

```python
# Imagem Binarizada (Preto e Branco)
# decidir se será preto ou branco
# valor maior que 128, vira 255 (branco)
# menor ou igual, vira 0 (preto)
# ponto médio
limite = 128

imagem_binaria = imagem_cinza.point(lambda p: 255 if p > limite else 0)
print("Imagem convertida para preto e branco (binarizada).")
```

```python
# Mostrar as três imagens na tela
# plt.subplots(1, 3) cria 1 linha com 3 espaços para imagens
fig, eixos = plt.subplots(1, 3, figsize=(15, 5))

# Exibe a primeira imagem (original)
eixos[0].imshow(imagem_original)
eixos[0].set_title('1. Imagem original')
eixos[0].axis('off')

# Exibe a segunda imagem (cinza)
# 'cmap=gray' mapa em tons de cinza
eixos[1].imshow(imagem_cinza, cmap='gray')
eixos[1].set_title('2. Imagem em níveis de cinza')
eixos[1].axis('off')

# Exibe a terceira imagem (binarizada)
eixos[2].imshow(imagem_binaria, cmap='gray')
eixos[2].set_title('3. Imagem Binarizada (P&B)')
eixos[2].axis('off')

# Ajusta espaçamento entre as imagens caso "espremidas/sobrepostas"
plt.tight_layout()

# Mostra as três imagens
plt.show()

print("As imagens estão na tela.")
```

Este notebook demonstra como transformar uma imagem em níveis de cinza e, em seguida, binarizá-la usando as bibliotecas `requests`, `PIL` (Pillow) e `matplotlib`.

Primeiro, importamos as bibliotecas necessárias: `requests` para baixar a imagem da internet, `PIL` para manipulação de imagens e `matplotlib` para exibição. Definimos a URL da imagem que será utilizada.

Baixamos a imagem da internet utilizando a biblioteca `requests`. O código inclui tratamento de erro para verificar se o download foi bem-sucedido.

Convertendo a imagem original para níveis de cinza. A função `convert('L')` da biblioteca PIL realiza essa transformação, onde 'L' significa luminância (escala de cinza).

Binarizamos a imagem em níveis de cinza. Definimos um limite (neste caso, 128). Pixels com valor de luminância maior que o limite se tornam brancos (255) e pixels com valor menor ou igual ao limite se tornam pretos (0).

Exibimos as três imagens (original, em níveis de cinza e binarizada) utilizando `matplotlib`. As imagens são exibidas lado a lado para comparação.
